#Operasi dalam phyton

latihan

```python
# 1. Hitung jumlah botol langsung di print().
print("Jumlah total botol:", 24 * 3)
```

```python
# 2. Simpan hasil perkalian (24 x 3) ke variabel total_jumlah_botol lalu tampilkan dengan print().
total_jumlah_botol = 24 * 3
print("Jumlah total botol:", total_jumlah_botol)
```

```python
# 3. Simpan angka 24 ke variabel banyak_botol_per_kardus dan angka 3 ke variabel jumlah_kardus,
#    lalu tampilkan perkaliannya langsung di print().
banyak_botol_per_kardus = 24
jumlah_kardus = 3
print("Jumlah total botol:", banyak_botol_per_kardus * jumlah_kardus)
```

```python
# 4. Simpan angka 24 ke variabel banyak_botol_per_kardus dan angka 3 ke variabel jumlah_kardus,
#    lalu simpan hasil perkaliannya ke variabel total_jumlah_botol, kemudian tampilkan dengan print().
banyak_botol_per_kardus = 24
jumlah_kardus = 3
total_jumlah_botol = banyak_botol_per_kardus * jumlah_kardus
print("Jumlah total botol:", total_jumlah_botol)
```
