#Aplikasi Menghitung Luas persegi Panjang

```python
Panjang = int(input("Masukkan Panjang dari Panjang (cm) : "))
Lebar = int(input(" Masukan Lebar dari persegi Panjang (cm): "))

Luas = Panjang * Lebar

print(f"\nLuas dari persegi Panjangnya adalah {Luas} cm^2.")
```

#Aplikasi Menghitung Volume Kubus

```python
sisi = int(input("masukan panjang sisi kubus (cm) : "))

Volume = sisi * sisi * sisi

print(f"\nVolume kubus adalah {Volume} cm^3.")
```
